```python
import os, io, gzip, zipfile, tarfile, sys, math, random
import numpy as np
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt

# For reproducibility
random.seed(42)
np.random.seed(42)

print(nx.__version__)
```

```python
# === Download the email-EuAll dataset from SNAP ===
import os
import urllib.request
import pandas as pd
import networkx as nx

# URL and local path
DATA_URL = "https://snap.stanford.edu/data/email-EuAll.txt.gz"
LOCAL_PATH = "data/raw_euAll.txt.gz"

# Create folder if it doesn't exist
os.makedirs("data", exist_ok=True)

# Function to download dataset
def download_dataset(url: str, to_path: str):
    if not os.path.exists(to_path):
        print(f"Downloading from {url} ...")
        urllib.request.urlretrieve(url, to_path)
        size = os.path.getsize(to_path) / (1024*1024)
        print(f"Saved to {to_path} ({size:.2f} MB)")
    else:
        print(f"Dataset already exists at {to_path}")

# Download dataset
download_dataset(DATA_URL, LOCAL_PATH)

# === Read dataset into pandas ===
edges = pd.read_csv(
    LOCAL_PATH,
    compression='gzip',   # file is gzipped
    sep='\t',             # some SNAP files use tabs; we’ll try whitespace
    names=["source", "target"],
    header=None,
    comment='#',          # skip any comment lines starting with #
    engine='python'
)

print(f"Dataset shape: {edges.shape}")
print(edges.head())

# === Build directed graph with NetworkX ===
G = nx.from_pandas_edgelist(
    edges,
    source="source",
    target="target",
    create_using=nx.DiGraph()
)

# Print basic graph info (NetworkX 3.x compatible)
print("Graph type:", type(G))
print("Number of nodes:", G.number_of_nodes())
print("Number of edges:", G.number_of_edges())
print("Is directed?", G.is_directed())
```

```python
import matplotlib.pyplot as plt
import numpy as np
from collections import Counter

# Compute degrees
in_degrees = [d for n, d in G.in_degree()]
out_degrees = [d for n, d in G.out_degree()]

# Count frequency of each degree
in_degree_count = Counter(in_degrees)
out_degree_count = Counter(out_degrees)

# Sort degrees
in_deg, in_freq = zip(*sorted(in_degree_count.items()))
out_deg, out_freq = zip(*sorted(out_degree_count.items()))

# Plot In-Degree Distribution
plt.figure(figsize=(8,6))
plt.scatter(in_deg, in_freq, alpha=0.6)
plt.xscale('log')
plt.yscale('log')
plt.xlabel("In-degree")
plt.ylabel("Frequency")
plt.title("In-degree Distribution (Log-Log) — email-EuAll")
plt.grid(True, which="both", linestyle="--", linewidth=0.5)
plt.show()

# Plot Out-Degree Distribution
plt.figure(figsize=(8,6))
plt.scatter(out_deg, out_freq, alpha=0.6)
plt.xscale('log')
plt.yscale('log')
plt.xlabel("Out-degree")
plt.ylabel("Frequency")
plt.title("Out-degree Distribution (Log-Log) — email-EuAll")
plt.grid(True, which="both", linestyle="--", linewidth=0.5)
plt.show()
```
